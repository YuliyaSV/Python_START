```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from shapely.geometry import LineString

def f(x):
    return np.power(x, 3) - 50*x

def g(x):
    return -np.power(x, 4) + 88 * np.power(x, 2) - 241


x = np.linspace(-15, 15, 100)

fig, ax = plt.subplots()

line1, = ax.plot(x, f(x),
                 label='f(x)')

line2, = ax.plot(x, g(x),
                 label='g(x)')



first_line = LineString(np.column_stack((x, f(x))))
second_line = LineString(np.column_stack((x, g(x))))
intersection = first_line.intersection(second_line)


px, py = LineString(intersection).xy

    
points = ax.plot(px, py,
                 'o', label='f(x) = g(x)')
    
ax.legend(loc='lower center')
plt.show()

for i in range(len(px)):
    print('x:' + str(px[i]) + ' y:' + str(py[i]) )
```
